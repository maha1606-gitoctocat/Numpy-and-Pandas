```python
# Pandas Series

import pandas as pd
marks = pd.Series(
    [95, 92, 89, 85, 80],
    index=['Rank1', 'Rank2', 'Rank3', 'Rank4', 'Rank5']
)

print(f"Marks Series:\n{marks}")
print(marks)

# Indexing and Slicing

print(f"1st Rank Student Mark: {marks.iloc[0]}")
print(marks.iloc[0])

print(f"Top 3 Ranks: {marks.loc[['Rank1', 'Rank2', 'Rank3']]}")
print(marks.loc[['Rank1', 'Rank2', 'Rank3']])

print(f"3rd Ranked Student Mark: {marks.iloc[2]}")
print(marks.iloc[2])

print(f"Marks Greater Than 90:\n{marks[marks > 90]}")
print(marks[marks > 90])

# Manipulating Series

marks['Rank1'] = 100

print("\nAfter Updating Rank1:")
print(marks)

marks = marks.drop('Rank5')

print("\nAfter Removing Rank5:")
print(marks)

cgpa = marks / 10

print(f"CGPA: {cgpa}")
print(cgpa)
```

```python
# Pandas DataFrame

transactions = pd.DataFrame({
    'TransactionID': [101,102,103,104,105,106,107,108,109,110],
    'ProductCategory': [
        'Electronics','Clothing','Electronics','Furniture',
        'Clothing','Electronics','Furniture','Clothing',
        'Furniture','Electronics'
    ],
    'Region': [
        'North','South','North','East','West',
        'North','East','West','South','North'
    ],
    'Amount': [200,150,300,450,200,250,300,180,350,400]
})

print(f"\nTransactions DataFrame:\n{transactions}")

# Data Exploration

print(f"\nHead:\n{transactions.head()}")

print(f"\nTail:\n{transactions.tail()}")

print(f"\nShape: {transactions.shape}")

print(f"\nColumn Names: {list(transactions.columns)}")

print(f"\nData Types:\n{transactions.dtypes}")

print(f"\nProductCategory and Amount:\n{transactions[['ProductCategory', 'Amount']]}")

print(f"\nLast 3 Columns:\n{transactions.iloc[:, -3:]}")

print(
    f"\nNorth Region and Amount > 200:\n"
    f"{transactions[(transactions['Region'] == 'North') & (transactions['Amount'] > 200)]}"
)

print(f"\nProduct Category Counts:\n{transactions['ProductCategory'].value_counts()}")

print(f"\nUnique Regions: {transactions['Region'].unique()}")

print(
    f"\nMean Amount by Region:\n"
    f"{transactions.groupby('Region')['Amount'].mean()}"
)

# DataFrame Manipulation

transactions.loc[
    transactions['TransactionID'] == 102,
    'Amount'
] = 165

print(f"\nUpdated Amount for TransactionID 102:\n{transactions}")

transactions['Discount'] = transactions['Amount'] * 0.10

print(f"\nAfter Adding Discount Column:\n{transactions}")

transactions = transactions[
    transactions['TransactionID'] != 109
]

print(f"\nAfter Removing TransactionID 109:\n{transactions}")

transactions.drop(
    columns=['Discount'],
    inplace=True
)

print(f"\nAfter Deleting Discount Column:\n{transactions}")
```
